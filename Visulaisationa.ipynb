```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph: sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
categories = ['Apples', 'Bananas', 'Oranges']
sales = [30, 45, 20]

plt.figure(figsize=(8, 6))
plt.bar(categories, sales)
plt.xlabel('Fruit')
plt.ylabel('Sales')
plt.title('Bar Chart: Fruit Sales')
plt.grid(axis='y')
plt.show()
```

```python
# Generate random data
data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram: Random Data Distribution')
plt.grid(axis='y')
plt.show()
```

```python
expenses = ['Rent', 'Groceries', 'Entertainment', 'Utilities']
amounts = [1000, 500, 200, 300]

plt.figure(figsize=(8, 6))
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart: Monthly Expenses')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, marker='o', color='b', label='Random Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot: Random Data Points')
plt.grid(True)
plt.legend()
plt.show()
```
