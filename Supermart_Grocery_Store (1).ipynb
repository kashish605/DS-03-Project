**Changing data type of Order Date from object data type to  date data type**

```python
df['Order Date']=pd.to_datetime(df['Order Date'], errors = 'ignore')
```

```python
#The order date data type has changed:
df.info()
```

```python
#We want to find total sales by category:
df1=df.groupby('Category')['Sales'].sum()
```

```python
df1.head()
```

```python
#We'll use Bar Graph to show Category Wise Sales:
df1.plot(kind='bar')
plt.title('Category Wise Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

Analysis:
Eggs, Meat & Fish has the highest sales, accounting for approximately 29.5% of the total sales.
Food Grains follows closely, contributing around 29.1% to the overall sales.
Beverages and Fruits & Veggies also play significant roles, with approximately 28.9% and 28.7% of the total sales, respectively.

Suggestions:
Given the strong performance of the Eggs, Meat & Fish category, consider expanding its product range or promoting related items.
Monitor the performance of Beverages and Fruits & Veggies, as they are close contenders. Adjust marketing strategies or pricing if needed.
Ensure consistent availability and quality for all categories to maintain customer satisfaction.

```python
#Extracting month and year from date for further analysis:
df['Month'] = pd.to_datetime(df['Order Date']).dt.strftime('%B')
df['year'] = df['Order Date'].dt.year
df['month_no'] = df['Order Date'].dt.month
```

```python
df.head(21)
```

```python
#Now we wnat to find the Yearly Sales:
yearly_sales = df.groupby('year')['Sales'].sum()
yearly_sales.head()
```

```python
plt.pie(yearly_sales, labels=yearly_sales.index, autopct='%1.1f%%')
yearly_sales.plot(kind='pie')
plt.title('Yearly Sales', fontsize = 14)
plt.show()
```

The year 2017 and 2018 had more than 50 percent of the total sales which implies the sales increase as the
year increases.

```python
# Step 1: Extract relevant columns:
city_sales = df[['City','Sales']]
city_sales.head()
```

```python
# Step 2: Calculate total sales per city:
total_sales = city_sales.groupby('City',as_index=False).sum()
total_sales.head()
```

```python
# Step 3: Sort the cities by sales
sorted_cities = total_sales.sort_values(by='Sales', ascending=False)
```

```python
# Step 4: Select the top 5 cities:
top_cities = sorted_cities.head()
top_cities
```

```python
#Step 5: Plot the bar chart:
plt.bar(top_cities.index, top_cities['Sales'])
plt.xlabel('City')
plt.ylabel('Sales')
plt.title('Top 5 Cities by Sales')
```

Kanayakumari and Vellore Cities contribute to the maximum sales.

```python
# Extract day, month, and year from 'Order Date'
df['month_no'] = df['Order Date'].dt.month
df['Month'] = pd.to_datetime(df['Order Date']).dt.strftime('%B')
df['year'] = df['Order Date'].dt.year
```

```python
monthly_sales = df.groupby('Month',as_index=False)['Sales'].sum()
```

```python
monthly_sales.head(12)
```

```python
month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]

# Convert the 'Month' column to a categorical type with the defined order
monthly_sales['Month'] = pd.Categorical(monthly_sales['Month'], categories=month_order, ordered=True)

monthly_sales_sorted = monthly_sales.sort_values(by='Month')

monthly_sales_sorted.head(12)
```

```python
plt.plot(monthly_sales_sorted['Month'], monthly_sales_sorted['Sales'], marker='o')
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(monthly_sales_sorted['Month'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```
