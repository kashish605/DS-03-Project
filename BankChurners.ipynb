Visualisations

```python
df1=df.groupby(['Customer_Age','Attrition_Flag']) ['CLIENTNUM'].count().reset_index()
df1
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x='Customer_Age',y='CLIENTNUM',hue='Attrition_Flag',data=df1)
```

```python
df2=df.groupby(['Attrition_Flag','Gender']) ['CLIENTNUM'].count().reset_index()
df2
```

```python
sns.barplot(x='Attrition_Flag',y='CLIENTNUM', hue='Gender',data=df2)
```

```python
df3=df.groupby(['Attrition_Flag','Education_Level']) ['CLIENTNUM'].count().reset_index()
df3
```

```python
plt.figure(figsize=(20,8))
sns.scatterplot(x='Attrition_Flag',y='CLIENTNUM',hue='Attrition_Flag',style='Education_Level',data=df3)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition_Flag', y='Total_Trans_Amt', data=df)
plt.title('Outliers in Total Transaction Amount')
plt.show()
```

```python
age_mean = df['Customer_Age'].mean()
age_median = df['Customer_Age'].median()
age_mode = df['Customer_Age'].mode()[0]
age_std = df['Customer_Age'].std()
print(f'Age - Mean: {age_mean}, Median: {age_median}, Mode: {age_mode}, Std Dev: {age_std}')
```
